### Data Cleaning

```python
print(df['Item Fat Content'].unique())
```

```python
df['Item Fat Content'] = df['Item Fat Content'].replace({'LF': 'Low Fat',
                                                         'low fat':'Low Fat',
                                                         'reg': 'Regular'})
```

```python
print(df['Item Fat Content'].unique())
```

### Bussiness Requirements

```python
### KPI'S Requirements
```

```python
## Total sales
total_sales = df['Total Sales'].sum()

# Average Sales
avg_sales = df['Total Sales'].mean()

# No of item sold
no_of_items_sold = df['Total Sales'].sum()

#Ratings
avg_ratings = df['Rating'].mean()

# Display
print(f"Total Sales: ${total_sales:,.0f}")
print(f"Average Sales: ${avg_sales:,.0f}")
print(f"No_Of_Items_sold : {no_of_items_sold :,.1f}")
print(f"Avg_Rating : {avg_ratings :,.1f}")
```

### CHART REQUIREMENT

#### Total sales by Fat Contenet

```python
sales_by_fat = df.groupby('Item Fat Content')['Total Sales'].sum()
plt.pie(sales_by_fat, labels = sales_by_fat.index,
        autopct = '%.1f%%',
        startangle = 90,
       colors = ["#f7b500", "#4caf50"])
plt.title('Total_Sales By Fat Content')
plt.axis('equal')

plt.show()
```

```python
### Total Sales by Item Type
```

```python
# Group data by Item Type and calculate total sales
sales_by_type = df.groupby('Item Type')['Total Sales'].sum().sort_values(ascending=False)

# Plot bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(sales_by_type.index, sales_by_type.values)

# Axis labels and title
plt.xticks(rotation=90)
plt.xlabel('Item Type')
plt.ylabel('Total Sales')
plt.title('Total Sales by Item Type')

# Add labels on top of bars
for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        f'{int(bar.get_height()):,}',   # formatted with commas, no decimals
        ha='center', va='bottom', fontsize=8
    )
plt.tight_layout()
plt.show()
```

```python
### Fat Content by Outlet for Total Sales
```

```python
# Grouped data
grouped = df.groupby(['Outlet Location Type', 'Item Fat Content'])['Total Sales'].sum().unstack()
grouped = grouped[['Regular', 'Low Fat']]

# Plot
ax = grouped.plot(kind='bar', figsize=(8, 5), title='Outlet Tier by Item Fat Content', color=['#4caf50','#f7b500'])
plt.xlabel('Outlet Location Tier')
plt.ylabel('Total Sales')
plt.legend(title='Item Fat Content')

# Add text labels on bars
for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=8)  # integer labels

plt.tight_layout()
plt.show()
```

```python
### Total Sales by Outlet Establishment
```

```python
sales_by_year = df.groupby('Outlet Establishment Year')['Total Sales'].sum().sort_index()

# Plot
plt.figure(figsize=(7, 5))
plt.plot(
    sales_by_year.index, 
    sales_by_year.values, 
    marker='o', 
    linestyle='-', 
    color='#4caf50',    # Line color (green)
    markerfacecolor='#f7b500',  # Marker color (yellow-orange)
    markeredgecolor='#4caf50'   # Marker border same as line
)

# Labels and title
plt.xlabel('Outlet Establishment Year')
plt.ylabel('Total Sales')
plt.title('Outlet Establishment')

# Add text labels
for x, y in zip(sales_by_year.index, sales_by_year.values):
    plt.text(x, y, f'{int(y):,}', ha='center', va='bottom', fontsize=8, color='#4caf50')

plt.tight_layout()
plt.show()
```

```python
### Sales by Outlet Size
```

```python
sales_by_size = df.groupby('Outlet Size')['Total Sales'].sum()
plt.figure(figsize=(4,4))
plt.pie(sales_by_size, labels = sales_by_size.index, autopct='%1.1f%%',
        startangle=90,colors = ["#f7b500", "#4caf50","skyblue"]
       )
plt.tight_layout()
plt.show()
```

```python
### Sales by Outlet Location
```

```python
sales_by_location = df.groupby('Outlet Location Type')['Total Sales'].sum().reset_index()
sales_by_location = sales_by_location.sort_values('Total Sales', ascending=False)

plt.figure(figsize=(8, 3)) # Smaller height, enough width
ax = sns.barplot(x='Total Sales', y='Outlet Location Type', data=sales_by_location, color='green')

plt.title('Total Sales by Outlet Location Type')
plt.xlabel('Total Sales')
plt.ylabel('Outlet Location Type')

plt.tight_layout() # Ensures layout fits without scroll
plt.show()
```
